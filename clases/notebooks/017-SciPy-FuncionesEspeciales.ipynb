# Funciones especiales con SciPy

_`SciPy` es un paquete que incluye una colección de algoritmos matemáticos y funciones construidas sobre el paquete `NumPy`. En este cuaderno nos vamos a centrar en las **funciones especiales**._

En primer lugar, __importemos__ lo paquetes que vamos a utilizar:

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
from scipy import special
```

```python
np.info(special)
```

```python
x = np.linspace(0,20,100)
for ii in range(5):
    plt.plot(x, special.jn(ii, x))
plt.grid(True)
```
